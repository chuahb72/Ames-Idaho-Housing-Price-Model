# <font color='green'>Property Size related variables</font>

##### <font color='green'>Both gr_liv_area and 1st_flr_sf are among the top 3 largest coeff. From earlier pair plots, they have show strong colinearity. We explore dropping 1st_flr_sf from the feature list</font> 

```python
size_features = ['gr_liv_area', '1st_flr_sf', 'total_bsmt_sf',
                   'bsmtfin_sf_1', '2nd_flr_sf', 'lot_area']
```

```python
corr_list = size_features[:]
corr_list.append('saleprice')

corr_set = train_set[corr_list]
sns.pairplot(corr_set, y_vars=size_features, x_vars=['saleprice'], height=5);
```

```python
# Establish size of figure.
plt.figure(figsize=(10,10))

#print(corr.shape)
# Set up mask to be "True" in the upper triangle.
mask = np.zeros_like(corr_set.corr())
mask[np.triu_indices_from(mask)] = True

# Plot our correlation heatmap, while masking the upper triangle to be white.
with sns.axes_style("white"):
    sns.heatmap(corr_set.corr(), mask=mask, vmax=.3, square=True, annot=True, cmap="YlGnBu")
```
